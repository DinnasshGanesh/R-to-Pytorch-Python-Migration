```python
import torch
import torch.nn as nn

class HydroCNN(nn.Module):
    """
    A 1D Convolutional Neural Network for Hydrological Time-Series Prediction.
    Designed to extract temporal features from climate data (Rain, Temp, Sun).
    """
    def __init__(self, window_size=30, num_features=3):
        super(HydroCNN, self).__init__()

        # Layer 1: Feature Extraction (Detecting storm events)
        self.conv1 = nn.Conv1d(in_channels=num_features, out_channels=32, kernel_size=3, padding=1)

        # Layer 2: Pattern Recognition (Detecting seasonal trends)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3, padding=1)

        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.2) # Prevents overfitting

        # Final output: Predicted Tank Storage Level
        self.fc = nn.Linear(64 * window_size, 1)

    def forward(self, x):
        # Input shape: (Batch, Features, Window)
        x = self.relu(self.conv1(x))
        x = self.relu(self.conv2(x))
        x = x.view(x.size(0), -1) # Flatten
        x = self.dropout(x)
        return self.fc(x)
```
